```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
data = pd.read_csv('macro.csv')
```

```python
data = data.drop(columns=['Time'])
```

```python
data.columns
columns = [ 'GDP', 'Consumption', 'GDP Implicit Price Deflator inflation',
       'CPI less food and energy', 'Total CPI', 'GPDI', 'Inventories',
       'WILL5000PRFC']
```

```python
for i in columns:
    data[f'{i} filtered'] = sm.tsa.filters.bkfilter(data[f'{i}'].dropna())
data.tail(50)
```

```python
data.columns
```

```python
for i in columns:
    fig, ax = plt.subplots()
    data[f'{i} filtered'].plot(ax=ax, style=['r--', 'b-'])
    plt.show()
```
